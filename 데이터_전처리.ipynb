```python
DataFrame = pd.read_csv('/content/drive/MyDrive/data/pandas/kpop_idols.csv')

DataFrame.iloc[:,1:11].head(50)
df = DataFrame.loc[(DataFrame['Height'] != 0)&(DataFrame['Weight'] != 0)].dropna(subset=['Group'],axis=0).reset_index(drop=True) #몸무개와 키의 값이 0이 아닌 값들을 가져옴
#dropna는 결측치를 제거해주는 것이다 즉 subset 특정 행을 선택하고 이거에 NaN 값을 없에 준다 axis=0 는 행 axis=1은 열이다
#reset_index(drop=True) drop은 인덱스로 세팅한 열을 DataFrame내에서 삭제할지 여부를 결정한다.

df

```

```python
#피벗 테이블
#피벗 테이블은 방대한 표(예: 데이터베이스,스프레드 시트 등등) 데이터를 요약하는 통계 표
# pd.pivot_table(df, index = '행 인덱스' , columns = '열 인덱스', values = '조회하고 싶은 값', aggfunc= '집계')

df.head()

pd.pivot_table(df,index='Group',columns='Country',values='Height',aggfunc='min')

pd.pivot_table(df,index='Group',columns='Country',values='Weight', aggfunc='mean')

pd.pivot_table(df,index='Group',columns='Country',values='Height',aggfunc='max')

pd.pivot_table(df,index='Group',columns='Country',values='Weight',aggfunc='sum')

pd.pivot_table(df,index='Group',columns='Country',values='Height',aggfunc='count')

pd.pivot_table(df,index='Group',columns='Country',values='Weight',aggfunc='std')


```

```python
#그룹 통계
#그룹으로 묶기는 피벗 테이블과 유사한 개념이다
#같은 값을 그룹으로 묶어서 분석할 때 사용한다

#df.groupby('그룹 구분').count() - 갯수
#df.groupby('그룹 구문').sum() - 합계 등등 이 있음

df.groupby('Group').mean(['Height'])

df.groupby('Group').sum()

df.groupby('Country').sum()

df.groupby('Country').mean()


df.groupby('Group')['Height'].mean()

df.groupby('Group')['Weight'].sum()



df.groupby(['Group','Country']).mean()

df.groupby(['Group'])['Stage Name'].max()


print(df.groupby(['Group','Country']).mean().reset_index() )
#인덱를 리셋해서 인덱스를 준다 


df.groupby(['Group'])['Country'].max().reset_index()



df.groupby('Group').min(['Height'])
```

```python
#그룹 데이터 확인
#.groups는 그룹을 확인할 수 있다 (G)I-DLE': [253, 254, 255]  253, 254, 255이 번호에  I-DLE의 정보가 있음을 알 수 있다
df.groupby('Group').groups 

df.groupby('Group').get_group('(G)I-DLE')  # 본인이 원하는 그룹의 값을 가지고 올 수 있다 

 
```

```python
df.groupby('Group').agg({'Height':'max','Weight':'min'})

df.groupby('Group').agg({'Height':'mean','Weight':'sum'})
```
