```python
df = pd.read_csv('/content/drive/MyDrive/data/pandas/kpop_idols.csv')

df = df.iloc[20:30]
df

df = df [['K. Stage Name','Date of Birth','Group','Height','Weight','Birthplace','Former Group']]

df
```

```python
#결측값 처리하기
#결측값이란 입력이 누락된 값,자료가 입력되지 않은 값을 이야기한다
#0의 값도 결측값이다


df = df.replace(0,np.NaN) 
#df = df.replace(0,np.NaN)
# replace는 왼쪽에 있는 값을 오른쪽에 있는 값으로 변경해주는 것이다
# np.NaN은 numpy 에서 제공해 주는 NaN 값이라고 생각하면 될 것같다

df
df.info()
```

```python
#결측 데이터 확인 
#isnull()
#notnull()



df.isnull() #null 값이 있는 것만 True(1) 아무 값이라도 있으면 False 
df.notnull() #어떤 값이라도 있으면 True 아무 값도 없으면 False


df.isnull().sum(axis=0) #0이 열 이고 1이 행이다
```

```python
#결측 데이터 제거

df.dropna() #행에 null 값이 있으면 삭제함

df.dropna(subset=['Group']) #그룹 열에 null 값이 있으면 그 칼럼의 행을 없앤다

df.dropna(subset=['Group','Weight','Weight'])  

df.dropna(subset=['Weight','Height']) # 이런식으로 subset을 사용해줄 수 있다


df = df.dropna(subset=['Group'])
df
```

```python
#결측 데이터를 치환할 수 있다
# fillna('대체값')

df['Weight'] = df['Weight'].fillna('min')

df.replace('min',np.NaN)

```

```python
#최빈값 빈도수를 나타내는 값이지요
#value_counts()
print(df['Group'].value_counts(),'\n')

df['Birthplace'].value_counts()

df['Group'].value_counts().index

df['Group'].value_counts().index[0]
```

```python
#인덱스가 깨지거나 할 때 reset_index() 로 인덱스 재정렬을 해준다

df = df.reset_index(drop=True) # drop=True는 이전 인덱스를 버린다는 뜻이다

df

```

```python
#중복된 행 찾기 
#df.duplicated()


print(df.duplicated()) #행단위로 중복 데이터 찾기
df[df.duplicated()]

df[df['Group'].duplicated()] #열 group 을 기준으로 중복되는 값을 출력해 준다

df[df['Birthplace'].duplicated()]
```

```python
#중복 값을 제거
# drop_duplicates()

df.drop_duplicates() # 행을 기준으로 중복되는 값을 지운다


df.drop_duplicates(subset=['Group']) # 그룹이라는 칼럼 즉 열을 기준으로 중복을 삭제한다


df.drop_duplicates(subset=['Birthplace']) # 고향이라는 칼럼 즉 열을 기준으로 중복을 삭제한다

```
