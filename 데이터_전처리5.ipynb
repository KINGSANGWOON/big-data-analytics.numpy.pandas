```python
# apply, lambda, map

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/pandas/kpop_idols.csv')
```

```python
df = df.loc[df['Gender'] == 'M'][0:5].append(df.head(5)).reset_index(drop=True)
df
```

```python
df = df[['Height','Weight','Gender']]
df
```

```python
# apply 함수는 구체적인 (로직)을 적용하고 싶을 경우 사용한다

def change_value(a):
  if a==0:
    a=170
  return a

df['Height'] = df['Height'].apply(change_value) # df['Height']의 값을 한 번씩 apply 가 지정해준 함수에 넣어서 돌린 후에 그 값을 Height 에 넣어준다


def change_value2(x):
  if x==63:
    x=60
  return x

df['Weight'] = df['Weight'].apply(change_value2)
```

```python
df['Weight']
```

```python
#lambda 함수
# apply에서 def와 정의된 함수와 동일한 기능을 한다
# 간단한 로직일 경우에는 lambda 함수를 사용한다
# lambda='입력변수':'리턴값' '조건문(옵션)'

df['Weight']=df['Weight'].apply(lambda x:60 if x==65 else x) # Weight의 값을 x로 받고 return 값은 60 그리고 조건은 만약 65일 경우 60 return 아니면 x값 그대로 출력

df
```

```python
df['Weight+Height'] = df.apply(lambda x:x['Weight']+x['Height'],axis=1)


```

```python
df['Weight+Height'] = df['Weight+Height'].astype('int')

df.info()
```

```python
#map 함수
#데이터 값을 특정한 값으로 mapping 한다

map_data = {'M':'남자'}

df['Gender'] = df['Gender'].map(map_data)

df
```
