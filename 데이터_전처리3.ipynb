```python
DataFrame = pd.read_csv('/content/drive/MyDrive/data/pandas/kpop_idols.csv')

df = DataFrame.iloc[:,1:11].tail(10)
```

```python
df = df[['Korean Name','Date of Birth','Height','Weight']]
df
```

```python
#자료형 변환
#astype('변환하려는 자료형') 으로 데이터 자료형을 변환할 수 있다
# int,float,str(object),datetime이 있다 

df.info()

df['Weight'] = df['Weight'].astype('str')  # 자료형 변환 후에 덮어씌우면 된다
df.info()

df['Weight'] = df['Weight'].astype('int')

df.info()
```

```python
df['Height'] = df['Height'].astype('str')

df['Height'] = df['Height'].astype('int')
df.info()
```

```python
#str->datetime 타입 변환 pd.datetime(df['칼럼명'])
#datetime은 날짜와 시간 데이터를 처리하기 위함이다
#datatime 이라는 라이브러리도 지원하지만 pandas 가 더 편하고 효율적이다

df['Date of Birth'][1592] #인덱스 값을 줘야하는 것을 잊지 말자

df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])

df['Date of Birth'][1592] # 자동적으로 년 월 일 시간 분 초를 인식하여 나타내준다

#df.info()


#데이트 타임 슬라이싱

print(df['Date of Birth'])

df['Date of Birth'].dt # .dt의 의미는 datetime에서 우리가 slicing을 할 수 있는 준비를 했다고 생각해 주면 된다

df['Date of Birth'].dt.year #년도만 뽑아냄

df['Date of Birth'].dt.month

df['Date of Birth'].dt.day


df['Date of Birth'].dt.dayofweek #요일
#df['Date of Birth'].dt.dayofweek 는 (월:0  ,화:1  ,수:2  ,목:3  ,금:4  ,토:5  ,일:6)


df['Year of Birth'] = df['Date of Birth'].dt.year
df['Month of Birth'] = df['Date of Birth'].dt.month
df['Day of Birth'] = df['Date of Birth'].dt.day

df

```

```python
#데이터 타임의 연산
# 시간을 더하고 빼는 건 3월을 더하는 것이 아닌 3개월이라는 간격을 더하고 빼야한다. 즉 이 자료형이 시간 간격(pd.Timedelta)여야 한다 (하나만 변경되도 된다)
# 즉 pd.to_datetime을 해준다고 연산이 되는 것이 아닌 timedelta 의 자료형으로 변환을 해줘야한다
# pd.to_timedelta(df['칼럼명'],unit='시간 간격 단위')
# 변환 단위는 일,시,분 등이 있다 (YEAR,MOUNTH 안되는 듯)


df['Day of Birth'] = pd.to_timedelta(df['Day of Birth'],unit='day')
#df['Day of Birth'] = pd.to_timedelta(df['Day of Birth'],unit='day')

df.info()

df['Date of Birth'] + df['Day of Birth'] # 계산 값을 보면 야무지게 월까지 올라가 있는 것을 알 수 있다

```
