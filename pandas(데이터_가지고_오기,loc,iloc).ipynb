```python
#iloc
#숫자 인덱스(integer index)을 통해 접근하는 방법 /  숫자를 기준으로 가지고 옴

df.iloc[[0,3,5]] #그냥 데이터의 순서에 따라서라고 생각하면 될 것 같다 아까 loc은 그 인덱스의 숫자를 맞춰 1584,1587,1589 이런식으로 해줘야 겠지만 이건 그냥 순서를 보고 결정해주면 된다

df.iloc[4,4] # 자신이 원하는 데이터를 간단하게 뽑아올 수 있는 듯하다

df.iloc[:,1]

df.iloc[:,1:4]
```

```python
#참 거짓 여부를 판단

df['Height']>160

df.loc[df['Height']==0]

```

```python
#isin() : 내가 정한 리스트 안에 있는 값들만 가져오고 싶을 때 사용하는 방법이다
df['Country'].isin(['South Korea'])

df.loc[df['Country'].isin(['South Korea'])]


df.loc[df['Stage Name'].isin(['Lily','Haewon'])]
```

```python
#행,열 추가
# ignore_index=True을 반드시 입력해 줘야함 행 추가 시에만 기존 행과 인덱스 값이 다르면 에러가 발생하기 때문에 ignore_index=True를 해줘야한다
#행 삭제 df.drop(index=['삭제 인덱스'])
#열 삭제 df.drop(columns=['삭제하려는 열 이름'])


new_data = {
    'Stage Name':'sangwoon',
    'Full Name':'limsangwoon',
    'Korean Name':'임상운',
    'K. Stage Name':'얌상운',
    'Date of Birth':'2004-05-30',
    'Group':'NMIXX팬카페 회장',
    'Country':'South Korea',
    'Second Country':'South Korea',
    'Height':'185',
    'Weight':'85'
}

df = df.append(new_data,ignore_index=True)
df
```
