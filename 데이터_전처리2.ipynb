```python
DataFrame = pd.read_csv('/content/drive/MyDrive/data/pandas/kpop_idols.csv')


df1 = DataFrame.iloc[:,1:11].head(5)
df2 = DataFrame.loc[0:5,['Korean Name','Instagram']]

print(df1)
print(df2)
```

```python
# join
# Inner join , Left Join , Right Join, Outer Join
#df1.join(df2,on='키 인덱스',how='조인 방법',sort='정렬 여부')
# 조인 방법 상관 없이 열은 그냥 추가 된다고 생각해라 
# 조인을 사용하기 위해서는 인덱스를 기준으로 조인을 하기 때문에
# 키로 지정할 열을 인덱스화 시켜주어야 합니다
# set_index을 해주면 된다

df2.set_index('Korean Name')

df1.join(df2.set_index('Korean Name'), on='Korean Name',how='left') # 인스타 아이디를 가지지 않는 테이블을 모두 가지고 있다
```

```python
df1.join(df2.set_index('Korean Name'), on='Korean Name',how='right')  # 인스타 아이디를 가진 테이블의 값은 모두 있다 right이기 때문에 
```

```python
df1.join(df2.set_index('Korean Name'),on='Korean Name',how='inner') # 두 테이블이 공통된 데이터만 나옴
```

```python
df1.join(df2.set_index('Korean Name'), on='Korean Name', how='outer') # 두 테이블의 모든 값을 전부 넣음


df1.join(df2.set_index('Korean Name'), on='Korean Name', how='outer', sort='Height')
```

```python
# Merge
# join은 index를 set_index를 통해서 지정해야하지만 
# Merge는 따로 인덱스를 지정해 주는 작업이 없어진다 
# 즉 value 기준으로 합치는 것이다
# Inner Merge, Left Merge , Right Merge , Outer Merge
# pd.merge(df1,df2, on='키 칼럼',how='조인 방법',sort='정렬 여부')


pd.merge(df1,df2,on='Korean Name', how='inner')

pd.merge(df1,df2, on='Korean Name', how='right')

pd.merge(df1,df2, on='Korean Name', how='left')

pd.merge(df1,df2,on='Korean Name', how='outer')
```

```python
# Concat
# 두개의 데이터 프레을 단순히 붙이는 방법
# 행이나 열 방향으로 데이터 프레임을 붙일 수 있다
#pd.concat([df1,df2],axis='0 or 1')
```

```python
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=0)

pd.concat([df1,df2],axis=0,join='inner') 
```
